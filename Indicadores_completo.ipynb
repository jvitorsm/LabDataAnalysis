```python
# Importação dos dados
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('analises_09.xlsx', parse_dates=True)
df
```

```python
# Quantidades de analises

qa = pd.DataFrame()

qa['PRODUTO'] = df['PRODUTO']
qa['TEOR'] = df['TEOR']
qa['TEOR'] = qa['TEOR'] * 100
qa['SOLUBILIDADE'] = df['SOLUBILIDADE']
qa['SOLUBILIDADE'] = qa['SOLUBILIDADE'] * 100
qa['PH'] = df['PH']
qa['UMIDADE'] = df['UMIDADE']
qa['UMIDADE'] = qa['UMIDADE'] *100
qa['Gran 2 mm'] = df['G. 2.0 mm'] * 100
qa['Gran 1 mm'] = df['G. 1.0 mm'] * 100
qa['Gran 0.71 mm'] = df['G. 0.71 mm'] * 100
qa['Pb']  = df ['Pb']
qa['Cd']  = df ['Cd']
qa = qa.fillna(0)

t = 0 
for i in qa['TEOR']:
  if i > 0:
    t+=1


s = 0
for i in qa['SOLUBILIDADE']:
  if i != 0:
    s+=1

 
p = 0
for i in qa['PH']:
  if i != 0:
    p+=1

u = 0
for i in qa['UMIDADE']:
  if i != 0:
    u+=1

g_2 = 0
for i in qa['Gran 2 mm']:
  if i != 0:
    g_2 += 1

g_1 = 0
for i in qa['Gran 1 mm']:
  if i != 0:
    g_1 += 1


g_07 = 0
for i in qa['Gran 0.71 mm']:
  if i != 0:
    g_07 += 1

pb = 0
for i in qa['Pb']:
  if i!= 0:
    pb += 1

cd = 0
for i in qa['Cd']:
  if i!= 0:
    cd += 1


quantidade_analise = t + s + p + u + g_2 + g_1 + g_07 + pb + cd

print(quantidade_analise)
```
